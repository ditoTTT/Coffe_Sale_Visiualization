```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/complete-pandas-tutorial/master/warmup-data/coffee.csv')
coffee = df.copy()
```

```python
coffee
```

```python
plt.style.use('fivethirtyeight')
fig, ax = plt.subplots(2, 1, figsize=(8,10))
bar_colors = ['#0471A6', '#FFBC0A']
coffee.pivot_table(values='Units Sold', columns='Coffee Type', index='Day', aggfunc='sum').plot.bar(ax=ax[0], title=('Coffee Sales Per Day'),color=bar_colors)
ax[0].set_title('Coffee Sales Per Day - Bar Chart')
ax[0].set_xlabel('Day')
ax[0].set_ylabel('Units Sold')

line_colors = ['#8F2D56', '#3F4531']
coffee.pivot_table(values='Units Sold', columns='Coffee Type', index='Day', aggfunc='sum').plot(ax=ax[1], title=('Coffee Sales Per Day'),linewidth=4, color=line_colors)
ax[1].set_title('Coffee Sales Per Day - Line Chart')
ax[1].set_xlabel('Day')
ax[1].set_ylabel('Units Sold')

plt.tight_layout()

plt.savefig('Visualization')
plt.show()
```
